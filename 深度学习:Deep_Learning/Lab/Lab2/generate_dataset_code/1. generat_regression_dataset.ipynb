```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# 参数设置
n_samples = 10000  # 总样本数
n_train = 7000  # 训练集样本数
n_test = 3000  # 测试集样本数
p = 500  # 特征维度
coef = 0.0056  # 系数
bias = 0.028  # 偏置
```

```python
# 生成特征数据
X = np.random.randn(n_samples, p)
```

```python
# 生成噪声
epsilon = np.random.randn(n_samples) * 0.01  # 可调整噪声幅度
```

```python
# 计算目标值 y
y = bias + np.dot(X, np.full(p, coef)) + epsilon
```

```python
# 划分训练集和测试集
X_train, X_test = X[:n_train], X[n_train:]
y_train, y_test = y[:n_train], y[n_train:]
```

```python
# 将数据保存为 Pandas DataFrame 以便查看
train_data = pd.DataFrame(X_train)
train_data['target'] = y_train

test_data = pd.DataFrame(X_test)
test_data['target'] = y_test
```

```python
# 保存数据的目录
save_path = r'C:\Users\37623\jupyter_test\Lab2_Nerual_Network\dataset\regression_data'
os.makedirs(save_path, exist_ok=True)
```

```python
# 保存训练集和测试集
train_data.to_csv(os.path.join(save_path, 'train_data.csv'), index=False)
test_data.to_csv(os.path.join(save_path, 'test_data.csv'), index=False)

print("Data saved successfully to /dataset/regression_data")
```

```python
# 显示训练集和测试集的部分数据
print("Training Data Sample:")
print(train_data.head())
print("\nTesting Data Sample:")
print(test_data.head())
```

```python
# 可视化：目标值 y 的分布
plt.figure(figsize=(10, 5))
plt.hist(y, bins=50, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribution of Target Values (y)')
plt.xlabel('Target Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 可视化：特征均值与目标值的关系（散点图）
# 这里选取 X 的第一列特征来展示与目标值的关系
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], y, alpha=0.5, color='green')
plt.title('Feature (X[:, 0]) vs Target (y)')
plt.xlabel('Feature X[:, 0]')
plt.ylabel('Target y')
plt.show()
```

# 说明

1. **生成特征**：使用 `np.random.randn` 生成标准正态分布的特征矩阵 `X`。
2. **生成噪声**：通过随机噪声项 `epsilon` 增加数据的随机性。
3. **计算目标值 \( y \)**：使用给定的线性方程 $$ y = \text{bias} + \text{np.dot}(X, \text{np.full}(p, \text{coef})) + \epsilon $$。
4. **划分数据集**：将生成的数据集分为训练集和测试集。
5. **输出**：使用 Pandas DataFrame 保存并输出前几行数据以便查看。
