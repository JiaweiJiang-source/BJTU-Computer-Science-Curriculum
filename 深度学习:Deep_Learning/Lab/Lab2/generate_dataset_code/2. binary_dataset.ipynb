```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# 参数设置
n_samples = 10000  # 每个数据集的总样本数
n_train = 7000  # 训练集样本数
n_test = 3000  # 测试集样本数
p = 200  # 特征维度
mean1 = 1  # 第一个数据集的均值
mean2 = -1  # 第二个数据集的均值
std_dev = 1  # 两个数据集的标准差（相同）
```

```python
# 生成第一个数据集（标签为0）
X1 = np.random.normal(loc=mean1, scale=std_dev, size=(n_samples, p))
y1 = np.zeros(n_samples)
```

```python
# 生成第二个数据集（标签为1）
X2 = np.random.normal(loc=mean2, scale=std_dev, size=(n_samples, p))
y2 = np.ones(n_samples)
```

```python
# 划分训练集和测试集
X1_train, X1_test = X1[:n_train], X1[n_train:]
y1_train, y1_test = y1[:n_train], y1[n_train:]

X2_train, X2_test = X2[:n_train], X2[n_train:]
y2_train, y2_test = y2[:n_train], y2[n_train:]
```

```python
# 合并两个数据集
X_train = np.vstack((X1_train, X2_train))
y_train = np.hstack((y1_train, y2_train))

X_test = np.vstack((X1_test, X2_test))
y_test = np.hstack((y1_test, y2_test))
```

```python
# 转换为 Pandas DataFrame 格式以便查看
train_data = pd.DataFrame(X_train)
train_data['target'] = y_train

test_data = pd.DataFrame(X_test)
test_data['target'] = y_test
```

```python
# 创建保存数据的目录
save_path = r'C:\Users\37623\jupyter_test\Lab2_Nerual_Network\dataset\binary_data'
os.makedirs(save_path, exist_ok=True)
```

```python
# 保存训练集和测试集
train_data.to_csv(os.path.join(save_path, 'train_data.csv'), index=False)
test_data.to_csv(os.path.join(save_path, 'test_data.csv'), index=False)

print("Data saved successfully to C:\\Users\\37623\\jupyter_test\\Lab2_Nerual_Network\\dataset\\binary_data")
```

```python
# 显示训练集和测试集的部分数据
print("Training Data Sample:")
print(train_data.head())
print("\nTesting Data Sample:")
print(test_data.head())
```

```python
# 可视化：两个类别样本的部分特征分布（这里选取前两个特征做散点图）
plt.figure(figsize=(10, 5))
plt.scatter(X1[:, 0], X1[:, 1], alpha=0.5, label='Class 0', color='blue')
plt.scatter(X2[:, 0], X2[:, 1], alpha=0.5, label='Class 1', color='red')
plt.title('Feature Distribution of Two Classes (Feature 0 vs Feature 1)')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.legend()
plt.show()
```

```python
# 可视化：第一个特征的分布直方图（验证均值和方差）
plt.figure(figsize=(10, 5))
plt.hist(X1[:, 0], bins=50, alpha=0.5, label='Class 0', color='blue', density=True)
plt.hist(X2[:, 0], bins=50, alpha=0.5, label='Class 1', color='red', density=True)
plt.title('Distribution of Feature 0 for Class 0 and Class 1')
plt.xlabel('Feature 0')
plt.ylabel('Density')
plt.legend()
plt.show()
```
